  - **Online Advertising Budget ($\text{Online}$):** The amount of money spent on online advertising campaigns (a continuous variable).

**Considering Meaningful Interactions:**

- **Interaction Effect:**
  - An interaction might need to be taken into account because the effectiveness of TV advertising may depend on the amount spent on online advertising and vice versa. This means the combined effect of TV and online advertising on sales might be different from the sum of their individual effects.

**Linear Models Without and With Interaction:**

1. **Without Interaction (Additive Model):**

   The model assumes that the effects of TV and online advertising on sales are independent and additive.

   **Equation:**

   $$
   \text{Sales} = \beta_0 + \beta_1 \times \text{TV} + \beta_2 \times \text{Online}
   $$

   - $\beta_0$: Intercept (baseline sales when no advertising is done).
   - $\beta_1$: Effect of TV advertising on sales.
   - $\beta_2$: Effect of online advertising on sales.

2. **With Interaction (Synergistic Model):**

   The model includes an interaction term to capture the effect of TV and online advertising working together.

   **Equation:**

   $$
   \text{Sales} = \beta_0 + \beta_1 \times \text{TV} + \beta_2 \times \text{Online} + \beta_3 \times (\text{TV} \times \text{Online})
   $$

   - $\beta_3$: Interaction effect between TV and online advertising.

**Using the Formulas to Make Predictions:**

- **Without Interaction:**

  - **Prediction Process:**
    - Plug in the values for $\text{TV}$ and $\text{Online}$ advertising budgets into the equation.
    - Calculate the predicted sales by summing the intercept and the products of each predictor with its coefficient.

  - **Interpretation:**
    - The effect of TV advertising on sales is constant, regardless of the online advertising budget.Conversely, The effect of online advertising on sales is constant, regardless of the TV advertising budget.
    - The total effect is simply the sum of the individual effects.

- **With Interaction:**

  - **Prediction Process:**
    - Plug in the values for $\text{TV}$ and $\text{Online}$ advertising budgets.
    - Calculate the interaction term ($\text{TV} \times \text{Online}$) and include it in the equation.
    - Sum all terms to get the predicted sales.

  - **Interpretation:**
    - The effect of one advertising medium on sales depends on the level of the other medium.
    - Captures synergistic effects where the combined impact is greater than the sum of individual effects.

**Difference Between Predictions:**

- **Additive Model:** Assumes independence between predictors; the impact of TV advertising on sales does not change with different levels of online advertising. Conversely, the impact of online advertising does not change with different levels of TV advertising budget.
- **Interaction Model:** Recognizes dependency between predictors; the impact of TV advertising varies depending on the online advertising budget. Conversely, the impact of online advertising varies depending on the TV advertising budget.

- **Mapping the Outcome Variable:**
  - Converted the binge eating frequency responses to a binary variable where "At least once a week" is mapped to 1 (indicating binge eating disorder), and all other responses are mapped to 0.

- **Mapping the Predictor Variable (Social Meals Frequency):**
  - **Categories Mapped to 1 (At least once a day):**
    - "1 time per day"

**Model Building:**

1. **Additive Model (No Interaction):**

````python
# Import Libraries
import pandas as pd
import statsmodels.formula.api as smf
import numpy as np
import plotly.express as px
from scipy import stats

# Load the data
data_url = "https://raw.githubusercontent.com/pointOfive/stat130chat130/main/CP/CSCS_data_anon.csv"
data = pd.read_csv(data_url, na_values=["9999", "", " ", "Presented but no response", "NA"])

# Keep only the columns we're interested in
columns_to_keep = [
    'NUTRITION_three_factor_eating_questionnaire_binge_eating_when_not_hungry',
    'LONELY_dejong_emotional_social_loneliness_scale_score',
    'NUTRITION_canadian_food_intake_screener_frequency_social_meals'
]
data = data[columns_to_keep]

# Prepare the Outcome Variable
# Map the outcome variable to binary
data['binge_eating_disorder'] = data['NUTRITION_three_factor_eating_questionnaire_binge_eating_when_not_hungry'].map({
    'At least once a week': 1,
    'Rarely': 0,
    'Never': 0,
    'Sometimes': 0
})

# Drop rows with missing outcome values
data = data.dropna(subset=['binge_eating_disorder'])

# Prepare the Predictor Variables
# Drop rows with missing loneliness scores
data = data.dropna(subset=['LONELY_dejong_emotional_social_loneliness_scale_score'])

# Define categories that represent eating meals with others at least once a day
daily_meals = ["1 time per day", "2 to 3 times per day"]

# Map the predictor variable to binary
def map_social_meals(value):
    if value in daily_meals:
        return 1
    elif pd.isna(value):
        return np.nan  # Keep NaN values to drop them later
    else:
        return 0

data['social_meals_daily'] = data['NUTRITION_canadian_food_intake_screener_frequency_social_meals'].apply(map_social_meals)

# Drop rows with missing predictor values
data = data.dropna(subset=['social_meals_daily'])

# Build the Logistic Regression Models
# Model 1: Additive Model (No Interaction)
formula_additive = 'binge_eating_disorder ~ LONELY_dejong_emotional_social_loneliness_scale_score + social_meals_daily'
model_additive = smf.logit(formula=formula_additive, data=data).fit()
print("Additive Model Summary:")
print(model_additive.summary())

# Model 2: Interaction Model (With Interaction)
formula_interaction = 'binge_eating_disorder ~ LONELY_dejong_emotional_social_loneliness_scale_score * social_meals_daily'
model_interaction = smf.logit(formula=formula_interaction, data=data).fit()
print("\nInteraction Model Summary:")
print(model_interaction.summary())

# Likelihood Ratio Test to assess the necessity of the interaction term
lr_stat = 2 * (model_interaction.llf - model_additive.llf)
p_value = stats.chi2.sf(lr_stat, df=1)  # df=1 because the interaction model has one additional parameter
print(f"\nLikelihood Ratio Test Statistic: {lr_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Visualization including lines fitted to observed data and simulated data

def plot_with_simulated_data(model, model_name):
    # Extract coefficients
    params = model.params
    intercept = params['Intercept']
    predictors = params.index.drop('Intercept')
    
    # Create a range of values for the continuous predictor
    loneliness_values = np.linspace(data['LONELY_dejong_emotional_social_loneliness_scale_score'].min(),
                                    data['LONELY_dejong_emotional_social_loneliness_scale_score'].max(), 100)
    
    # Prepare data for both levels of the binary predictor
    plot_data = pd.DataFrame({
        'LONELY_dejong_emotional_social_loneliness_scale_score': np.concatenate([loneliness_values, loneliness_values]),
        'social_meals_daily': [0]*100 + [1]*100
    })
    
    # Calculate the linear predictor (log-odds) for observed data
    if 'LONELY_dejong_emotional_social_loneliness_scale_score:social_meals_daily' in predictors:
        coef_loneliness = params['LONELY_dejong_emotional_social_loneliness_scale_score']
        coef_social_meals = params['social_meals_daily']
        coef_interaction = params['LONELY_dejong_emotional_social_loneliness_scale_score:social_meals_daily']
        
        plot_data['linear_pred_observed'] = (
            intercept
            + coef_loneliness * plot_data['LONELY_dejong_emotional_social_loneliness_scale_score']
            + coef_social_meals * plot_data['social_meals_daily']
            + coef_interaction * plot_data['LONELY_dejong_emotional_social_loneliness_scale_score'] * plot_data['social_meals_daily']
        )
    else:
        coef_loneliness = params['LONELY_dejong_emotional_social_loneliness_scale_score']
        coef_social_meals = params['social_meals_daily']
        
        plot_data['linear_pred_observed'] = (
            intercept
            + coef_loneliness * plot_data['LONELY_dejong_emotional_social_loneliness_scale_score']
            + coef_social_meals * plot_data['social_meals_daily']
        )
    
    # Simulate random noise and generate simulated data points
    np.random.seed(0)
    num_points = 200
    loneliness_sim = np.random.uniform(data['LONELY_dejong_emotional_social_loneliness_scale_score'].min(),
                                       data['LONELY_dejong_emotional_social_loneliness_scale_score'].max(), num_points)
    loneliness_sim += np.random.normal(0, 0.5, num_points)  # Add noise
    
    # Assign random values for the binary predictor
    social_meals_sim = np.random.choice([0, 1], size=num_points)
    
    # Simulate outcome variable using the same model coefficients
    sim_data = pd.DataFrame({
        'LONELY_dejong_emotional_social_loneliness_scale_score': loneliness_sim,
        'social_meals_daily': social_meals_sim
    })
    
    # Compute the linear predictor for simulated data
    if 'LONELY_dejong_emotional_social_loneliness_scale_score:social_meals_daily' in predictors:
        sim_data['linear_pred_simulated'] = (
            intercept
            + coef_loneliness * sim_data['LONELY_dejong_emotional_social_loneliness_scale_score']
            + coef_social_meals * sim_data['social_meals_daily']
            + coef_interaction * sim_data['LONELY_dejong_emotional_social_loneliness_scale_score'] * sim_data['social_meals_daily']
        )
    else:
        sim_data['linear_pred_simulated'] = (
            intercept
            + coef_loneliness * sim_data['LONELY_dejong_emotional_social_loneliness_scale_score']
            + coef_social_meals * sim_data['social_meals_daily']
        )
    
    # Fit a new model to the simulated data
    sim_data['binge_eating_disorder'] = sim_data['linear_pred_simulated']  # Since we're approximating linear regression
    sim_model = smf.ols(formula='binge_eating_disorder ~ LONELY_dejong_emotional_social_loneliness_scale_score * social_meals_daily', data=sim_data).fit()
    
    # Extract coefficients from the simulated model
    sim_params = sim_model.params
    sim_intercept = sim_params['Intercept']
    sim_predictors = sim_params.index.drop('Intercept')
    
    # Calculate the linear predictor for the simulated model
    plot_data['linear_pred_simulated'] = None  # Initialize the column
    
    if 'LONELY_dejong_emotional_social_loneliness_scale_score:social_meals_daily' in sim_predictors:
        sim_coef_loneliness = sim_params['LONELY_dejong_emotional_social_loneliness_scale_score']
        sim_coef_social_meals = sim_params['social_meals_daily']
        sim_coef_interaction = sim_params['LONELY_dejong_emotional_social_loneliness_scale_score:social_meals_daily']
        
        plot_data['linear_pred_simulated'] = (
            sim_intercept
            + sim_coef_loneliness * plot_data['LONELY_dejong_emotional_social_loneliness_scale_score']
            + sim_coef_social_meals * plot_data['social_meals_daily']
            + sim_coef_interaction * plot_data['LONELY_dejong_emotional_social_loneliness_scale_score'] * plot_data['social_meals_daily']
        )
    else:
        sim_coef_loneliness = sim_params['LONELY_dejong_emotional_social_loneliness_scale_score']
        sim_coef_social_meals = sim_params['social_meals_daily']
        
        plot_data['linear_pred_simulated'] = (
            sim_intercept
            + sim_coef_loneliness * plot_data['LONELY_dejong_emotional_social_loneliness_scale_score']
            + sim_coef_social_meals * plot_data['social_meals_daily']
        )
    
    # Visualization
    # Plot the best fit lines for observed data
    fig = px.line(
        plot_data,
        x='LONELY_dejong_emotional_social_loneliness_scale_score',
        y='linear_pred_observed',
        color='social_meals_daily',
        labels={
            'LONELY_dejong_emotional_social_loneliness_scale_score': 'Loneliness Score',
            'linear_pred_observed': 'Linear Predictor (Observed Data)',
            'social_meals_daily': 'Social Meals Daily'
        },
        title=f'{model_name} Approximation with Observed and Simulated Data'
    )
    
    # Plot the best fit lines for simulated data
    for meal_freq in [0, 1]:
        subset = plot_data[plot_data['social_meals_daily'] == meal_freq]
        fig.add_scatter(
            x=subset['LONELY_dejong_emotional_social_loneliness_scale_score'],
            y=subset['linear_pred_simulated'],
            mode='lines',
            name=f'Simulated Fit - Social Meals {meal_freq}',
            line=dict(dash='dash')
        )
    
    # Add simulated data points
    fig.add_scatter(
        x=sim_data['LONELY_dejong_emotional_social_loneliness_scale_score'],
        y=sim_data['binge_eating_disorder'],
        mode='markers',
        marker=dict(color='grey', opacity=0.5),
        name='Simulated Data Points'
    )
    
    fig.show(renderer='png')
    
    return sim_data

# Plot for the additive model
plot_with_simulated_data(model_additive, 'Additive Model')

# Plot for the interaction model
sim_data=plot_with_simulated_data(model_interaction, 'Interaction Model')

print(model_interaction.predict(sim_data))
````

"The model only explains 17.6% of the variability in the data" does not necessarily contradict with "many of the coefficients are larger than 10 while having strong or very strong evidence against the null hypothesis of 'no effect'". A low R-squared value does not conflict with low p-values for the coefficients because they address different aspects of the model.

The p-value of a coefficient in the fitted model characterizes the strength of evidence against the null hypothesis that the coefficient (representing the relationship between the outcome and a predictor) is zero in the theoretical multiple linear regression (MLR) model.

It reflects our level of confidence that a true association exists between the outcome and the predictor in the theoretical multiple regression model, based on the observed data.

In contrast, the R-squared represents the proportion of variation in the outcome explained by the predictors.

It measures how well our model, which represents the combined effect of all associations, fits the data. It shows if the observed data actually aligns with the theoretical relationships suggested by our coefficients. 

This measure is independent of the confidence we have in the estimated theoretical relationships (as indicated by the p-values). It can be thought of as "How tightly the data points distribute around the model due to random noise"

Therefore, it is possible for our coefficients to have low p-values (indicating strong evidence of an effect) while the model has a low R-squared value (indicating that the effects explain only a small portion of the variability in the data). This suggests that, although we are confident in the existence of linear associations between the predictors and the outcome in the theoretical multiple linear regression model, the actual observed data fit those associations poorly, perhaps due to high variability / a large error term.

### 6. Work with a ChatBot to understand how the *model4_linear_form* (*linear form* specification of  *model4*) creates new *predictor variables* as the columns of the so-called "design matrix" *model4_spec.exog* (*model4_spec.exog.shape*) used to predict the *outcome variable*  *model4_spec.endog* and why the so-called *multicollinearity* in this "design matrix" (observed in *np.corrcoef(model4_spec.exog)*) contribues to the lack of "out of sample" *generalization* of *predictions* from *model4_fit*; then, explain this consisely in your own works<br>

When we specified interaction terms in the model4_linear_form, Python needed to create a new predictor variable for each interaction term and add these to the columns of the design matrix, represented by model4_spec.exog.

The design matrix is a structured table where each row corresponds to an observation, and each column represents every term in our model’s equation, including original predictors and all specified interaction terms. It's through using this table that Python was able to fit a least-squares model that predicts the outcome values ,which are stored in a tuple assigned to `model4_spec.endog`.

Using this table,Python can display the values of each predictor (including interaction terms) for each observation in the dataset. This allows us to measure collinearity by checking the linear correlations between the values of two predictors across different observations.

Multicollinearity arises when values in a full collection of multiple pairs of columns (predictor variables) are linearly intercorrelated with each other, meaning that there are many predictors whose values vary together. In such cases, the model fails to distinguish the individual relaitionships between these each predictor and the outcome, making the estimates of their coefficients unstable and potentially unreliable. 

As a result, predictions based on these coefficients may (but not necessarily) perform poorly on new data, indicating low generalizability.

In other words, the model might capture random associations that exist only in the training data due to the specific combinations of correlated predictors. However, these correlations might not hold in new test data, causing the model to perform well on training data but poorly on unseen data. The model is essentially overfitting, where it learns patterns that do not generalize.

Imagine trying to figure out which ingredient makes your cake sweet when you always use both sugar and honey in the same ratio. It’s hard to tell whether the sweetness comes from the sugar, the honey, or both. While this may work well when making the same cake repeatedly, it becomes difficult to predict the sweetness of a new recipe where the sugar and honey ratio is different.

````python
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
from sklearn.model_selection import train_test_split
import plotly.express as px
import plotly.graph_objects as go

# Assuming pokeaman_train is your full dataset
# Define all possible categories for Generation, Type 1, and Type 2
all_generations = pokeaman_train['Generation'].unique()
all_type1 = pokeaman_train['Type 1'].unique()
all_type2 = pokeaman_train['Type 2'].unique()

# Convert categorical variables to pandas Categorical with predefined categories
pokeaman_train['Generation'] = pd.Categorical(pokeaman_train['Generation'], categories=all_generations)
pokeaman_train['Type 1'] = pd.Categorical(pokeaman_train['Type 1'], categories=all_type1)
pokeaman_train['Type 2'] = pd.Categorical(pokeaman_train['Type 2'], categories=all_type2)

# Initialize variables to store performance metrics
reps = 100  # Number of repetitions
in_sample_Rsquared = np.zeros(reps)
out_of_sample_Rsquared = np.zeros(reps)

# Define the linear form specification
model5_linear_form = 'HP ~ Attack + Defense + Speed + Legendary + Q("Sp. Def") + Q("Sp. Atk") + C(Generation) + C(Q("Type 1")) + C(Q("Type 2"))'

for i in range(reps):
    # Split the data into training and testing sets (50-50 split)
    train_data, test_data = train_test_split(pokeaman_train, train_size=0.5)
    
    # Ensure categorical variables in test_data have the same categories as training data
    test_data['Generation'] = pd.Categorical(test_data['Generation'], categories=all_generations)
    test_data['Type 1'] = pd.Categorical(test_data['Type 1'], categories=all_type1)
    test_data['Type 2'] = pd.Categorical(test_data['Type 2'], categories=all_type2)
    
    # Fit the model on the training data
    model_fit = smf.ols(formula=model5_linear_form, data=train_data).fit()
    
    # Record the in-sample R-squared
    in_sample_Rsquared[i] = model_fit.rsquared
    
    # Predict on the testing data
    predictions = model_fit.predict(test_data)
    
    # Calculate the out-of-sample R-squared
    out_of_sample_Rsquared[i] = np.corrcoef(test_data['HP'], predictions)[0, 1] ** 2

# Create a DataFrame with the Collected Metrics
df_performance = pd.DataFrame({
    "In Sample R-squared": in_sample_Rsquared,
    "Out of Sample R-squared": out_of_sample_Rsquared
})

# Visualize the Results
fig = px.scatter(df_performance, 
                 x="In Sample R-squared", 
                 y="Out of Sample R-squared",
                 title="In Sample vs Out of Sample R-squared for Model5",
                 labels={
                     "In Sample R-squared": "In Sample R²",
                     "Out of Sample R-squared": "Out of Sample R²"
                 })

# Add a y=x reference line
fig.add_trace(go.Scatter(
    x=[0, 1], y=[0, 1],
    mode='lines',
    name='y = x',
    line=dict(color='red', dash='dash')
))

# Show the plot
fig.show(renderer='png')
````

My results display the distribution of In-sample R² against Out-of-sample R² for a large number of models trained using a random 50% of the data and tested on the other 50%, evaluating their generalizability. I observe that Out-of-sample R² is generally lower than In-sample R², which shows that for our dataset, it is challenging for the models to achieve optimal generalizability.

I also observe that there seems to be a negative association between In-sample R² and Out-of-sample R²: as the model fits the training data better, it tends to generalize more poorly to the rest of the data.

The purpose of this demonstration is to highlight the importance of balancing in-sample predictive performance with out-of-sample generalizability. This balance ensures that models not only perform well on the data they were trained on but also maintain their predictive accuracy on new, unseen data.
