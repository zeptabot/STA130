```python
import numpy as np  # Import numpy for random selection

# Initialize variables
all_door_options = [1, 2, 3]  # List representing the three doors (1, 2, 3)
my_door_choice = 1            # Your initial choice, fixed at door 1 for the simulation
i_won = 0                     # Counter to track how many times you win
reps = 100000                 # Number of times to repeat the simulation

# Loop through the simulation reps times
for i in range(reps):
    # Randomly select the winning door from the list of doors
    secret_winning_door = np.random.choice(all_door_options)
    
    # The host reveals a goat door (not the winning door and not your initial choice)
    goat_door_reveal = np.random.choice([door for door in all_door_options if door != secret_winning_door and door != my_door_choice])
    
    # The remaining door after your choice and the host's revealed goat door
    remaining_door = [door for door in all_door_options if door != my_door_choice and door != goat_door_reveal][0]
    
    # You switch your choice to the remaining door
    my_door_choice = remaining_door
    
    # Check if the new choice is the winning door
    if my_door_choice == secret_winning_door:
        i_won += 1  # If you win, increment the win counter

# Calculate the win ratio (i.e., how often you win when switching)
win_ratio = i_won / reps

# Print the win ratio
print(f"Win Ratio: {win_ratio}")
```
