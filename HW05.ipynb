The p-value is the probability of obtaining a test statistic as or more extreme than the one we actually observed, assuming the null hypothesis is true. We can visualize this as the area under the curve of the sampling distribution under the null hypothesis that is further away from the hypothesized parameter value than our observed test statistic.

In this hypothetical world, we describe what the population would look like if the null hypothesis were indeed true. 

Then, we could repeatedly take samples from this hypothetical population and calculate the test statistic each time. By calculating the proportion of values in this sampling distribution that are as or more extreme than our observed test statistic, we obtain the p-value. In that world, the population has a parameter specified by the null hypothesis. If we were to repeatedly take random samples from this population, we could calculate the test statistic for each sample. This would give us a distribution of possible test statistic values in this world, forming "the sampling distribution under the null hypothesis" , or "the null distribution".

Therefore, "Data as or more extreme" would mean statistics further away from the hypothesized population parameter than our observed statistic on this null distribution. For instance, if our null hypothesis claims no difference in average test scores between two groups, a difference in average scores larger than what we actually observed could be considered "more extreme."

Hence, the p-value can be interpreted as the proportion of data further away from the hypothesized population parameter than our observed statistic on the null distribution. This suggests our observed statistic is more "ridiculous" and "unlikely" compared to what is expected under the null hypothesis. 

However, since we know that the observed statistic is real and did occur, we are led to consider it's actually the null hypothesis itself that is more "ridiculous" and "unlikely" to be true. 

This might lead us to reject the null hypothesis and consider alternative explanations for our findings. Hence, a smaller p-value makes the null hypothesis look more ridiculous.

### 5. G√ºnt√ºrk√ºn (2003) recorded how kissing couples tilt their heads. 80 out of 124 couples, or 64.5% tilted their heads to the right. Simulate a **p-value** using a "50/50 coin-flipping" model for the assumption of the **null hypothesis** $H_0$ that the population of humans don't have left or right head tilt tendencies when kissing, and use the table below to determine the level of evidence we have against $H_0$. <br>


|p-value|Evidence|
|-|-|
|$$p > 0.1$$|No evidence against the null hypothesis|
|$$0.1 \ge p > 0.05$$|Weak evidence against the null hypothesis|
|$$0.05 \ge p > 0.01$$|Moderate evidence against the null hypothesis|
|$$0.01 \ge p > 0.001$$|Strong evidence against the null hypothesis|
|$$0.001 \ge p$$|Very strong evidence against the null hypothesis|

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(10)

# Parameters
sample_size = 124
observed_proportion = 80 / 124
simulations = 10000
null_prob = 0.5

# Simulating 10,000 trials of 124 couples tilting right or left using a fair coin (null hypothesis)
simulated_right_tilts = np.random.binomial(sample_size, null_prob, simulations)

# Convert simulated right tilts to proportions
simulated_proportions = simulated_right_tilts / sample_size

# Calculate two-tailed p-value (extreme cases on either side)
extreme_proportion = np.abs(simulated_proportions - 0.5) >= np.abs(observed_proportion - 0.5)
p_value = np.mean(extreme_proportion)

# Visualization
plt.figure(figsize=(10, 6))
plt.hist(simulated_proportions, bins=30, edgecolor='black', alpha=0.7)
plt.axvline(observed_proportion, color='red', linestyle='dashed', linewidth=2, label=f'Observed proportion: {observed_proportion:.3f}')
plt.axvline(2 * 0.5 - observed_proportion, color='red', linestyle='dashed', linewidth=2)

# Annotating the observed and extreme thresholds
plt.annotate('Observed Right Tilt Proportion', xy=(observed_proportion, 400), xytext=(observed_proportion + 0.02, 450),
             arrowprops=dict(facecolor='black', shrink=0.05))
plt.annotate('Extreme Left Tilt Proportion', xy=(2 * 0.5 - observed_proportion, 400), xytext=(2 * 0.5 - observed_proportion - 0.05, 450),
             arrowprops=dict(facecolor='black', shrink=0.05))

# Labels and title
plt.title('Sampling Distribution of Proportion of Right Tilts (under null hypothesis)')
plt.xlabel('Proportion of Right Tilts')
plt.ylabel('Frequency')
plt.legend()

# Show p-value on the plot
plt.figtext(0.4, 0.85, f'P-value: {p_value:.4f}', fontsize=12, color='red')

# Render the plot
plt.savefig("right_tilts.png", format="png")
plt.show()

print("based on the strength of evidence table given above, we have 'Strong' evidence against the null hypothesis.")
```

### Fisher's Tea Experiment

## We are interested in studying whether or not the students of STA130 possess the ability to identify whether the milk or tea was poured first into a cup of tea.

This experiment is SIMILAR to, however DIFFERENT from the Fisher experiment, where the goal was to determine whether Dr. Muriel Bristol the individual could correctly identify the pouring order of milk and tea.

In the Fisher experiment, the population was (hypothetically) every cup of tea Bristol would ever drink, and the parameter of interest was the proportion of the cups of tea for which she could correctly state which was poured first. The null hypothesis was that Bristol was just randomly guessing, and had no ability to distinguish the teas. The sample data was that 8 out of 8 cups was correctly identified by Bristol.

In our experiment, the context is different:

The population is all the students in STA130.

Parameter of interest (p): The proportion of students who could correctly state whether milk or tea was poured first.

Sample statistic ($\bar{p}$): The proportion of 49 students who correctly identified the pouring order out of 80 sampled students, 49/80. Or approximately, 0.6125.

Since we are only interested in scenarios where students show tea-distinguishing ability, in which case they would perform better than the "random" proportion of 50%, we will be conducting a one-tailed hypothesis test. We are only interested in data even larger than our observed statistic.

Therefore:

Formal null hypotheses $H_0$: p<=0.5

Explaination: The students have no real ability to identify which whether milk or tea was poured first. This means that, on average, No more than 50% of the students would correctly identify the pouring order purely by chance.

Alternative hypothesis $H_A$: p>0.5

This implies that students possess some genuine ability to correctly identify the pouring order, therefore performing better than random guessing. Hence, the true proportion is greater than 0.5.

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt

# Ensure plots display in the notebook
%matplotlib inline

# Step 1: Set seed for reproducibility
np.random.seed(42)

# Step 2: Define sample size and observed data
n = 80  # Total number of students in the sample
observed_successes = 49  # Number of students who correctly identified the pouring order
observed_proportion = observed_successes / n

print("=== Hypothesis Testing for STA130 Tea Experiment ===\n")
print(f"Sample size (n): {n}")
print(f"Number of students who correctly identified the pouring order: {observed_successes}")
print(f"Observed sample proportion (pÃÇ): {observed_proportion:.4f}")
print("We select the alpha-significance level of 0.05 as our p-value threshold for determining statistical significance.")

# Step 3: State the null and alternative hypotheses
print("\nNull Hypothesis (H0): p ‚â§ 0.5")
print("Alternative Hypothesis (H_A): p > 0.5")
print("\nWe assume that students are guessing, so the population parameter is 0.5.")

# Step 4: Perform the simulation
num_simulations = 10000  # Number of simulations
print(f"\nSimulating {num_simulations} samples using the assumed population parameter...")

# For each simulation, generate n trials with p=0.5 and compute the sample proportion
simulated_successes = np.random.binomial(n, 0.5, num_simulations)
simulated_proportions = simulated_successes / n

# Step 5: Calculate the p-value
# Since we are performing a one-tailed test (H_A: p > 0.5),
# we compute the proportion of simulations where the simulated sample proportion
# is greater than or equal to the observed proportion

# Count how many simulated proportions are greater than or equal to the observed proportion
count_extreme = np.sum(simulated_proportions >= observed_proportion)

# Calculate the p-value
p_value = count_extreme / num_simulations

print(f"\nNumber of simulations where simulated proportion ‚â• observed proportion: {count_extreme}")
print(f"Calculated p-value: {count_extreme} / {num_simulations} = {p_value:.4f}")

# Step 6: Visualize the sampling distribution
print("\nGenerating visualization of the null sampling distribution...")

plt.figure(figsize=(10, 6))
plt.hist(simulated_proportions, bins=20, edgecolor='black', alpha=0.7)
plt.axvline(observed_proportion, color='red', linestyle='dashed', linewidth=2)
plt.text(observed_proportion + 0.002, plt.ylim()[1]*0.9, 'Observed proportion', color='red')
plt.title('Null Sampling Distribution of Sample Proportions')
plt.xlabel('Sample Proportion')
plt.ylabel('Frequency')
plt.savefig("sampling_distribution.png", format="png")
plt.show()

# Step 7: Findings and Discussion
print("\n=== Findings and Discussion ===")
print(f"The number of simulations where the simulated proportion ‚â• observed proportion ({observed_proportion:.4f}) is {count_extreme}.")
print(f"Thus, the p-value is calculated as {count_extreme} divided by {num_simulations}:")
print(f"p-value = {count_extreme} / {num_simulations} = {p_value:.4f}")
print("\nThis p-value represents the probability of obtaining a sample proportion")
print(f"as large as or larger than {observed_proportion:.4f} under the null hypothesis.")
print("If the p-value is less than the chosen significance level (Œ± = 0.05),")
print("we would reject the null hypothesis in favor of the alternative hypothesis.")

# Step 8: Conclusion regarding the Null Hypothesis
alpha = 0.05  # Significance level
if p_value < alpha:
    print(f"\nConclusion:")
    print(f"Since the p-value ({p_value:.4f}) is less than the significance level (Œ± = {alpha}),")
    print("we reject the null hypothesis.")
    print("There is sufficient evidence to suggest that the proportion of STA130 students")
    print("who can correctly identify the pouring order is greater than 0.5.")
else:
    print(f"\nConclusion:")
    print(f"Since the p-value ({p_value:.4f}) is greater than the significance level (Œ± = {alpha}),")
    print("we fail to reject the null hypothesis.")
    print("There is insufficient evidence to suggest that the proportion of STA130 students")
    print("who can correctly identify the pouring order is greater than 0.5.")
```
